## <u>Understanding data types</u>
<p>Data may be classified into <b>continuous</b>(numeric and infinite) or <b>categorical</b>(discrete and finite). Pandas does not group data into the two formats but uses the following technical metrics:</p>

|Data type|Numpy/Pandas object|Pandas string name|Notes|
|:--|:--|:----------------------|:------------------------|
|Boolean|np.bool|bool|Stored as a single byte|
|Integer|np.int|int|Default of 64 bits. Unsigned integers are also available as np.uint|
|Float|np.float|float|Default of 64 bits|
|Complex|np.complex|complex|Rarely seen in data analysis|
|Object|np.object|O, object|Strings in Python objects eg. Tuples, lists, dictionaries, etc|
|Datetime|np.datatime64, pd.Timestamp|datetime64|Specific moment in time with nanosecond precision|
|Timedelta|np.timedelta64, pd.Timedelta|timedelta64|Amount of time from days to nanoseconds|
|Categorical|pd.categorical|category|Specific to Pandas only. Useful for object columns with relatively few unique values|

## <u>Chaining Series methods together</u>
<p>In Python, every variable is an object and all objects have attributes and methods that refer to them to return more objects. This sequential invocation of methods using the dot notation is referred to as <b>method chaining</b>. The backslash(\) sign is used to write the chained code in several lines. A parenthesis can also be used to wrap the whole expression. To improve readability, you can placeeach method directly under the dot above it</p>

```python
#The head method is commonly appended to the chain to suppress a long output.
#For shorter chains, there is no need to place each method on a different line:
director.value_counts().head(3)
```

```python
#A commonly used chain method is this one:
actor_1_facebook_likes.isnull().sum()
```

```python
actor_1_facebook_likes.notnull().sum()
```

```python
#All non-missing values should be integers, all missing values should be filled with zeros then be of integer data type:
actor_1_facebook_likes.fillna(0)\
                      .astype(int)\
                      .head()
```

```python
#We can also take the mean of the Series to get the percentage of values that are missing:
actor_1_facebook_likes.isnull().mean()
```

```python
#Parenthesis can be used as follows:
(actor_1_facebook_likes.fillna(0)
                 .astype(int)
                 .head())
```

<p>The method chaining method has its downfalls such as:
<ul>
<li>Challenges in debugging</li> 
<li>None of the intermediate outputs is stored in a variable</li>

## <u>Creating and deleting columns</u>

```python
#Creating a column and filling it with zeros. Created columns will go to the end of the dataset:
movie['has_seen'] = 0
```

```python
movie.head()
```

```python
#Adding up the columns with actor and director facebook likes and assign them to a new column:
movie['actor_director_facebook_likes'] = \
    (movie['actor_1_facebook_likes'] +
    movie['actor_2_facebook_likes'] +
    movie['actor_3_facebook_likes'] +
    movie['director_facebook_likes'])
```

<p>This dataset contains missing values. When numeric columns are added to one another as in the preceding step, pandas defaults 
missing values to zero. But, if all values for a particular row are missing, then pandas keeps the total as missing as well</p>

```python
movie['actor_director_facebook_likes'].isnull().sum()
```

```python
movie.actor_director_facebook_likes.fillna(0)
movie.actor_director_facebook_likes.isnull().sum()
```

```python
#The above code syntax can be replaced with this one below:
movie['actor_director_facebook_likes'] = \
movie['actor_director_facebook_likes'].fillna(0)
```

```python
movie['actor_director_facebook_likes'].isnull().sum()
```

```python
movie.head()
```

```python
#Check whether all values in the cast_total_facebook_likes are greater than the actor_total_facebook_likes:
movie['is_cast_likes_more'] = movie['cast_total_facebook_likes'] >= movie['actor_director_facebook_likes']
```

```python
movie['is_cast_likes_more'].all()
```

```python
movie['is_cast_likes_more'].notnull().sum()
```

```python
#Calculation of the total percentage of the cast_total_facebook_likes that come from actor_total_facebook_likes:
movie['pct_actor_cast_like'] = \
(movie['actor_director_facebook_likes'] /
movie['cast_total_facebook_likes'])
```

```python
movie['pct_actor_cast_like']
```

```python
#Let us validate that the min and max of this column fall between 0 and 1:
(movie['pct_actor_cast_like'].min(), movie['pct_actor_cast_like'].max())
```
