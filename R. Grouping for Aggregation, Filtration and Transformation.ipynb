## <u>Introduction</u>
<p>Fundamentally, data analysis involves the splitting of of data into independent groups before carrying out calculations on
each group. This strategy is sometimes referred to as <b>split-apply-combine</b>. This notebook covers the <b>groupby</b>
method to group data in any imaginable way and apply any function to each group before returning a single dataset. Hadley Wickham described the term split-apply-combine as a data analysis pattern of breaking up data into independent manageable chunks  to independently apply functions to these chunks and then combining the result back together. More details can be found in his <a href = 'http://bit.ly/2isFuL9'>paper</a>. Many methods can be chained in this groupby method thus unleashing its power. The result of the computation will be a <b>DataFrameGroupBy</b> or <b>SeriesGroupBy</b> but for simplicity they will be referred to as <b>groupby</b></p>

## <u>Defining aggregation</u>
<p>Aggregation is the process whereby a sequence of many inputs are summarized into a single value output. For instance, finding the maximum value of a column or summing all its values can be considered to be aggregation. Many values are therefore converted into a single value. Most aggregations have two components:</p>
<ul>
    <li><b>Aggregating columns </b>->Items whose values will be aggregated</li>
    <li><b>Aggregating functions </b>->How the aggregations will occur</li>
</ul>

<h2><u>Removing MultiIndex after grouping</u></h2>

## <u>Customizing an aggregation function</u>
<p>Sometimes, you may need to write down your own customized user-defined function that does not exist in Pandas or Numpy. Here
, we will use the college dataset to calculate the mean and standard deviation of the undergraduate student population per race
. We will then use this information to find the maximum number of standard deviations from the mean that any single population
value is per state.</p>

```python
#Read in the college dataset
college = pd.read_csv('data/college.csv')
```

```python
college.head()
```

```python
college.columns
```

find the mean and standard deviation of the undergraduate population by state:

```python
college.groupby('STABBR').agg({'UGDS' : ['mean', 'std']}).round(0).head()
```

```python
#An alternative syntax can be as follows:
college.groupby('STABBR')['UGDS'].agg(['mean', 'std']).round(0).head()
```

<p>We were not looking for the mean and standard deviation for the college dataset but rather the maximum  number of standard 
deviations away from the mean for any one institution. To do this, we subtract the mean undergraduate population by state from
each institution and then divide by the standard deviation. This will standardize the undergraduate population for each group.
We will take the maximum absolute value of these scores to find one that is furthest away from the mean. Pandas has no function
capable of this. Instead, we need to create a custom function.</p>

```python
def max_deviations(s):
    std_score = (s - s.mean()) / s.std()
    return std_score.abs().max()
```

```python
#We pass the function directly to the agg method to complete the aggregation:
max_sds = college.groupby('STABBR', as_index = False)['UGDS'].agg(max_deviations)\
                                 .round(1)\
                                  .head()
max_sds
```

```python
#The function can also be applied to multiple columns.
#The function only works with numeric columns:
college.groupby('STABBR')['UGDS', 'SATVRMID', 'SATMTMID']\
       .agg(max_deviations).round(1).head()
```

```python
#Customized functions can also be used with predefined functions:
max_dev_col = college.groupby(['STABBR', 'RELAFFIL']) \
        ['UGDS', 'SATVRMID', 'SATMTMID'] \
        .agg([max_deviations, 'mean', 'std']).round(1).head()
max_dev_col
```

```python
#You can advance on the cleanliness of the above data:
level0 = max_dev_col.columns.get_level_values(0)
level0
```

```python
level1 = max_dev_col.columns.get_level_values(1)
level1
```

```python
max_dev_col.columns = level0 + '_' + level1
```

```python
max_dev_col.head()
```

```python
last_result = max_dev_col.reset_index()
```

## <u>Customizing aggregation functions with *args and **kwargs</u>
<p>When writing user-defined customized aggrefation functions, Pandas passes the aggregating columns one at a time as a 
Series. You may need to pass several arguments to your function than just the Series itself. To do this, you need to be aware 
of the ability of Python to pass an arbitrary number of arguments to these functions. Let us now review the signature of the
agg method with help from the inspect module</p>

```python
grouped = college.groupby(['STABBR', 'RELAFFIL'])
grouped.head(5)
```

```python
import inspect
inspect.signature(grouped.agg)
```

<p>The argument *args allow you to pass an arbitrary number of non-keyword arguments to your customized aggregation function. 
Similarly, **kwargs allows you to pass an arbitrary number of keyword arguments. In this procedure, we will build a customized function for the college dataset that finds the percentage of schools by state and religious affiliation that have an undergraduate population between two values</p>

```python
college.head()
```

```python
#Define a function to return the percentage of schools with an undergraduate population between 1000 and 3000:
def pct_between_1_3k(s):
    return s.between(1000, 3000).mean()
```

```python
#Calculating the percentage by state and religious affiliation:
college.groupby(['STABBR', 'RELAFFIL'])['UGDS']\
       .agg(pct_between_1_3k).head(11)
```

```python
#The function is good but what if the user requires a customized lower and upper bound?
def pct_between(s, low, high):
    return s.between(low, high).mean()
```

```python
#Subject the new function to the agg method along with the upper and lower bounds:
#This is where *args and **kwargs apply wheen the non-keyworda arguments are passed
college.groupby(['STABBR', 'RELAFFIL'])['UGDS'].agg(pct_between, 1000, 10000).head(11)
```

```python
college.groupby(['STABBR', 'RELAFFIL'])['UGDS']\
       .agg(pct_between, 1000, 10000)\
       .head(11)           
```

```python
#The above code can also be written in the syntax below:
college.groupby(['STABBR', 'RELAFFIL'])['UGDS']\
       .agg(pct_between, low = 1000, high = 10000)\
       .head(11)
```

```python
#For ease of understanding, use the arguments in the order used when the function was being created
#Keyword and non-keyword arguments can also be mixed using the syntax below:
college.groupby(['STABBR', 'RELAFFIL'])['UGDS']\
       .agg(pct_between, 1000, high = 10000)\
       .head(11)
```

```python
#Technically, when agg is called, all the non-keyword arguments get collected into a tuple named args. 
#All the keyword arguments get collected into a dictionary named kwargs.
#Pandas does not accept additional arguments when using multiple aggregation functions together:
#The code below will result in an error:
#college.groupby(['STABBR', 'RELAFFIL'])['UGDS'] \
#       .agg(['mean', pct_between], low=100, high=1000)
#They can all be done procedurally one after the other
```

```python
#A function that calculates the total percentage of minority students and returns True than the user-defined threshold:
def check_minority(df, threshold):
    minority_pct = 1 - df['UGDS_WHITE']
    total_minority = (df['UGDS'] * minority_pct).sum()
    total_ugds = df['UGDS'].sum()
    total_minority_pct = total_minority / total_ugds
    return total_minority_pct > threshold
```

## <u>Calculating weighted mean SAT score per state with apply</u>
<p>The <b>groupby</b> function has four methods: <b>agg, filter, transform</b> and <b>apply</b>. The agg method returns a 
scalar value, filter returns a boolean and transform returns a Series. The apply method returns either a scalar value, Series
or a DataFrame of any shape thus making it flexible. In this recipe, we will calculate of the weighted average of the math and
verbal SAT scores per state from the college dataset. We weight the scores by the population of the undergraduate students per
school.</p>

```python
#Load the college dataset:
college = pd.read_csv('data/college.csv')
college.head()
```

```python
subset = ['UGDS', 'SATMTMID', 'SATVRMID']
```

```python
#Dropping missing values in accordance to specific colums
college2 = college.dropna(subset = subset)
college2.shape
```

```python
college.shape
```

```python
#The data left is still enough
#Creating a function to calculate the weighted average of just SAT math scores
def weighted_math_average(df):
    weighted_math = df['UGDS'] * df['SATMTMID']
    return int(weighted_math.sum() / df['UGDS'].sum())
```

```python
#This will result in a scalar value for each group
college2.groupby('STABBR').apply(weighted_math_average).head()
```

```python
#Subjecting the predefined function to the agg method:
#The function will apply even to other columns that do not require aggregation
#If the function is subjected to the only two columns, it will result in an error
#college2.groupby('STABBR').agg(weighted_math_average).head()
```

```python
#Simultaneous calculation of geometric, harmonic, arithmetic and weighted means
#The scipy.stata function is used to calculate the geometric and harmonic means
#The function will take the DataFrame name as its only argument
#It will create indexes that will be used to house the four types of means
#Numpy can be used to calculate the arithmetic and weighted means
#You will then join the indexes to the initialized methods through a for loop via indexing
#The rows are called inside the for loop to combine the indexes and the outputs of the four functions
#The other features can be initialized outside of the for loop
#All the data is returned as integers
from scipy.stats import gmean, hmean
def calculate_means(df):
    df_means = pd.DataFrame(index=['Arithmetic', 'Weighted', 'Geometric', 'Harmonic'])
    cols = ['SATMTMID', 'SATVRMID']
    for col in cols:
        arithmetic = df[col].mean()
        weighted = np.average(df[col], weights=df['UGDS'])
        geometric = gmean(df[col])
        harmonic = hmean(df[col])
        df_means[col] = [arithmetic, weighted, geometric, harmonic]
    df_means['Count'] = len(df)
    return df_means.astype(int)
```

```python
#Applying the function in the above cell to the specified rows:
#Apply method works with non-missing data only
college2.groupby('STABBR').apply(calculate_means).head(12)
```

## <u>Grouping continuous variables</u>
<p>Grouping requires discrete data with repeating values. If there are no repeated values, the output categories might each
have one column thus not being useful. Continuous numeric colums have few repeated values thus are not used to form groups. If
we transform continuous data to discrete data by placing each value to a bin, rounding them or any other form of mapping, then 
grouping would make sense. In this recipe, we will use the flights dataset to discover the distribution of airlines for
different distances travelled. For instance, we can find the airline that makes the most flights between 500 and 1000 miles.
To do  this, we use the Pandas function <b>cut</b> to discretize the distance of each flight flown</p>

```python
flights.head()
```

```python
#rCreation of five discrete bins of the distance feature
bins = [-np.inf, 200, 500, 1000, 2000, np.inf]
#You can pass an integer to the bins method to split according to the number of that integer
#Placing negative values is also catered for in this code. Values outside this range will not be considered
#Not placing -np.inf will make the negative values not to be considered
cuts = pd.cut(flights['DIST'], bins = bins)
cuts.head()
```

```python
#Counting the  number of flights per bin:
cuts.value_counts()
```

```python
#Formation of groups on cuts via the groupby method and subjection to value counts and normalization:
#SkyWest(OO) has 33% of flights less than 200 miles but only 16% of flighs between 200 and 500 miles
flights.groupby(cuts)['AIRLINE'].value_counts(normalize = True)\
                               .round(3)\
                               .head(15)
```

```python
#Informative string labels can also be created using the cut function
labels=['Under an Hour', '1 Hour', '1-2 Hours', '2-4 Hours', '4+ Hours']
cuts2 = pd.cut(flights['DIST'], bins = bins, labels = labels)
cuts2.head()
```

```python
#The unstack method transposes the inner index level to columns names
flights.groupby(cuts2)['AIRLINE'].value_counts(normalize = True)\
                                 .round(3)\
                                 .unstack()\
                                 .style.highlight_max(axis = 1)
```

```python
def max_streak(s):
    s1 = s.cumsum()
    return s.mul(s1).diff().where(lambda x: x < 0) \
            .ffill().add(s1, fill_value=0).max()
```

```python
flights.sort_values(['MONTH', 'DAY', 'SCHED_DEP']) \
       .groupby(['AIRLINE', 'ORG_AIR'])['ON_TIME'] \
       .agg(['mean', 'size', max_streak]).round(2).head()
```
